# RFM Analysis

### Importing libraries

```python
#importing libraries
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

### Importing data

```python
#importing data
cust=pd.read_csv('data.csv', encoding='ISO-8859-1')
```

```python
#checking data head
cust.head()
```

```python
#checking data info
cust.info()
```

```python
cust['InvoiceDate']=pd.to_datetime(cust['InvoiceDate'])
cust.info()
```

### Calculating recency

```python
cust_recency=cust.groupby('CustomerID',as_index=False)['InvoiceDate'].max()
recent_date=cust_recency['InvoiceDate'].max()
cust_recency.rename(columns={'InvoiceDate':'RecentDate'},inplace=True)
cust_recency['Recency']=cust_recency['RecentDate'].apply(lambda x: (recent_date-x).days)

cust_recency.head()
```

### Counting frequency

```python
cust_frequency=cust.groupby('CustomerID',as_index=False)['InvoiceDate'].count()
cust_frequency.rename(columns={'InvoiceDate':'Frequency'},inplace=True)
cust_frequency.head()
```

### Calculate monetary

```python
cust['Total']=cust['Quantity']*cust['UnitPrice']
cust_monetary=cust.groupby('CustomerID',as_index=False)['Total'].sum()
cust_monetary.rename(columns={'Total':'Monetary'},inplace=True)
cust_monetary.head()
```

### Merging all the data frames

```python
rf_df = cust_recency.merge(cust_frequency, on='CustomerID')
cust_rfm=rf_df.merge(cust_monetary,on='CustomerID')
cust_rfm.head()
```

### Ranking Customerâ€™s based upon their recency, frequency, and monetary score

```python
cust_rfm['R_rank'] = cust_rfm['Recency'].rank(ascending=False)
cust_rfm['F_rank'] = cust_rfm['Frequency'].rank(ascending=True)
cust_rfm['M_rank'] = cust_rfm['Monetary'].rank(ascending=True)
# normalizing the rank of the customers
cust_rfm['R_rank_norm'] = (cust_rfm['R_rank']/cust_rfm['R_rank'].max())*100
cust_rfm['F_rank_norm'] = (cust_rfm['F_rank']/cust_rfm['F_rank'].max())*100
cust_rfm['M_rank_norm'] = (cust_rfm['F_rank']/cust_rfm['M_rank'].max())*100
 
cust_rfm.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)
 

cust_rfm.head()
```

### Calculating RFM score

```python
cust_rfm['RFM_Score'] = 0.15*cust_rfm['R_rank_norm']+0.28 * cust_rfm['F_rank_norm']+0.57*cust_rfm['M_rank_norm']
cust_rfm['RFM_Score'] *= 0.05
cust_rfm = cust_rfm.round(2)
cust_rfm[['CustomerID', 'RFM_Score']].head(7)
```

### Rating Customer based upon the RFM score
rfm score >4.5 : Top Customer
4.5 > rfm score > 4 : High Value Customer
4>rfm score >3 : Medium value customer
3>rfm score>1.6 : Low-value customer
rfm score<1.6 :Lost Customer

```python


cust_rfm['Rfm rating']=np.where(cust_rfm['RFM_Score'] > 4.5, "Top Customers" ,(np.where( cust_rfm['RFM_Score'] > 4, "High value Customer", (np.where( cust_rfm['RFM_Score'] > 3,  "Medium Value Customer",np.where(cust_rfm['RFM_Score'] > 1.6,'Low Value Customers', 'Lost Customers'))))))
cust_rfm.head(5)
```

## Visualizing the amount of customers on the basis of their rating

```python
cust_explode=[0.01,0.01,0.01,0.01,0.01]
plt.pie(cust_rfm['Rfm rating'].value_counts(),labels=cust_rfm['Rfm rating'].value_counts().index,autopct='%.0f%%',explode=cust_explode)
plt.legend(title = "Rfm rating:",loc='upper right',bbox_to_anchor=(1,0))
plt.show()
```
